# Eyesuite export cleaning

```python
import csv

rows_out = []
header = True
with open('eyesuite_export.csv', 'r') as csvfile:
    reader = csv.reader(csvfile, delimiter=";")
    i = 0
    for row in reader:
        
        if header:
            length = len(row)
            if length != 865:
                row.extend([None,]*(865 - length))            
                rows_out.append(row)
            header = False
        
        if row[4] == "G":
            length = len(row)
            if length != 865:
                row.extend([-1,]*(865 - length))

            print(len(row))
            rows_out.append(row)

with open('eyesuite_export_001_onlyG.csv', 'w') as csvfile:
    writer = csv.writer(csvfile, delimiter=',', quoting=csv.QUOTE_MINIMAL)

    for row in rows_out:
        writer.writerow(row)
```

```python
import pandas as pd

# read the file created from heyex db matching the VFs with OCTs
df = pd.read_csv('002_heyex_export.csv')
df.dropna(subset=['dcm_path'], inplace=True)
```

```python
df.drop(df.columns[:2], axis=1, inplace=True)
```

```python
df.head()
```

```python
df['test_date'] = pd.to_datetime(df.testdate)
df['vf_test_day'] = df.test_date.dt.date
df['vf_test_time'] = df.test_date.dt.time

df.drop(['test_date', 'testdate'], axis=1, inplace=True)
```

```python
# discard double VFs on the same eye on the same day, keeping the latest
idx = df.groupby(['patient_id', 'image_laterality', 'vf_test_day'])['vf_test_time'].transform(max) == df['vf_test_time']
```

```python
df_heyex = pd.read_csv('004_heyex_export_cleaned_moreinfo.csv')
df_heyex.drop(df_heyex.columns[:2], axis=1, inplace=True)
```

```python
original_len = len(df_heyex)
df_heyex = df_heyex.sort_values('dcm_acquisition_date', ascending=False)
df_heyex.drop_duplicates('image_hash', inplace=True)
print(original_len - len(df_heyex))
```

```python
df_heyex.groupby(['patient_id', 'image_laterality', 'vf_test_day'])['dcm_path'].count().value_counts()
```

After deleting duplicates, we are left with ~20k OCTs, of which 17144 are already unique. The remaining 3000 need to be filtered further, as the VFs have more than one corresponding OCT.

158 volumes are not acceptable, i.e. 49-slice volumes with 768 or 1024 pixels. See next line

```python
# 158 volumes are not acceptable, i.e. 49-slice volumes with 768 or 1024 pixels
df_heyex = df_heyex[~((df_heyex['slices'] == 49) & (df_heyex['columns'].isin([768, 1024])))]
df_heyex = df_heyex[~((df_heyex['slices'] == 61) & (df_heyex['columns'] == 512))]
df_heyex = df_heyex[~((df_heyex['slices'] == 49) & (df_heyex['columns'] == 384))]
```

```python
len(df_heyex)
```

```python
df_heyex['grid_width'] = df_heyex['columns'] * df_heyex['pxl_spacing_y']
df_heyex['grid_height'] = df_heyex['slices'] * df_heyex['slice_thickness']
```

```python
df_heyex.describe()
```

We now set priorities to volumes:
- If volume has 61 slices, priority 1 (2nd choice)
- If volume has 49 slices and covers the 6mm grid in both directions, priority 0 (1st choice)
- If volume has 49 slices and doesn't cover the 6mm grid, priority 2 (3rd choice)

```python
df_heyex[df_heyex['slices'] == 61]['grid_height'].sort_values()
```

```python
df_heyex[df_heyex['slices'] == 61]['grid_height'].hist(bins=50)
```

```python
df_heyex[df_heyex['slices'] == 61]['grid_width'].hist(bins=50)
```

```python
df_heyex.drop([18830], axis=0, inplace=True)
```

```python
df_heyex.describe()
```

```python
df_heyex['priority'] = 1 # set all to 61-slice priority, since we have verified that they all cover the grid (5.97 mm is the minimum)

fortynine_covering = (df_heyex['slices'] == 49) & (df_heyex.grid_height >= 6) & (df_heyex.grid_width >= 6)
fortynine_not_covering = (df_heyex['slices'] == 49) & ((df_heyex.grid_height < 6) | (df_heyex.grid_width < 6))
# verified that the sum is equal to the length of the df
# assert 4810 + 10449 + 5049 == len(df_heyex)

df_heyex.loc[fortynine_covering, 'priority'] = 0
df_heyex.loc[fortynine_not_covering, 'priority'] = 2
```

```python
df_only61 = df_heyex[df_heyex['slices'] == 61]

idx = df_only61.groupby(['patient_id', 'image_laterality', 'vf_test_day'])['dcm_acquisition_date'].transform(max) == df_only61['dcm_acquisition_date']
df_only61 = df_only61[idx]

df_only61.groupby(['patient_id', 'image_laterality', 'vf_test_day'])['dcm_path'].count().value_counts()
```
